```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import warnings
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

```

```python
# 데이터 불러오기
data = pd.read_csv('한국교통안전공단_지역 간 대중교통 시간대별 버스 이용인원_20221231.csv', encoding='cp949')
data.head()
```

```python
# 시간대별 평균 이용인원 구하기
data['평균_고속'] = data[['평일(고속)', '토요일(고속)', '일요일(고속)']].mean(axis=1)
data['평균_시외'] = data[['평일(시외)', '토요일(시외)', '일요일(시외)']].mean(axis=1)

# 클러스터링에 사용할 특성 선택
X = data[['평균_고속', '평균_시외']]
```

```python
# 엘보우 방법을 사용하여 최적의 클러스터 수 찾기
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# 클러스터 수에 따른 WCSS 그래프 그리기
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# 최적의 클러스터 수 결정
optimal_k = 2  

# K-means 클러스터링 수행
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
kmeans.fit(X)

# 클러스터링 결과 확인
labels = kmeans.labels_
data['Cluster'] = labels

# 클러스터링 결과 시각화
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k):
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['평균_고속'], cluster_data['평균_시외'], label=f'Cluster {cluster}')
plt.xlabel('avg use(Express Bus)')
plt.ylabel('avg use(intercity bus)')
plt.title('Clustering of Bus Usage Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 클러스터링 결과의 실루엣 점수 계산
silhouette_avg = silhouette_score(X, labels)

print("전체 데이터셋의 실루엣 점수:", silhouette_avg)
```
